```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create distribution plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Sales distribution
sns.histplot(data['Sales'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Sales')
axes[0].set_xlabel('Sales')

# Profit distribution
sns.histplot(data['Profit'], bins=30, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Profit')
axes[1].set_xlabel('Profit')

# Discount distribution
sns.histplot(data['Discount'], bins=30, kde=True, ax=axes[2])
axes[2].set_title('Distribution of Discount')
axes[2].set_xlabel('Discount')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot to show relationship between Sales and Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=data, hue='Discount', palette='viridis', alpha=0.6)
plt.title('Relationship between Sales and Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend(title='Discount')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Sales', data=data, palette='viridis')
plt.title('Boxplot of Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# Top 10 Products by Sales
plt.figure(figsize=(14, 10))
top_products = data.groupby('Product Name')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False).head(10)
sns.barplot(x='Sales', y='Product Name', data=top_products, palette='viridis')
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```

```python
# Profit by Segment
plt.figure(figsize=(12, 8))
segment_profit = data.groupby('Segment')['Profit'].sum().reset_index().sort_values(by='Profit', ascending=False)
sns.barplot(x='Profit', y='Segment', data=segment_profit, palette='viridis')
plt.title('Total Profit by Segment')
plt.xlabel('Total Profit')
plt.ylabel('Segment')
plt.show()
```

```python
# Boxplot of Sales by Region
plt.figure(figsize=(12, 8))
sns.boxplot(x='Region', y='Sales', data=data, palette='viridis')
plt.title('Boxplot of Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```
